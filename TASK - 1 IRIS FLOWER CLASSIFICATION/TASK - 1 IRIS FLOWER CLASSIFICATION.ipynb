# Data Modelling

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns = ['Species'])
y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.4)
y_test
```

# Decision Tree : ID3 decision tree classifier


The ID3 algorithm, short for Iterative Dichotomiser 3, derives its name from its iterative and dichotomizing approach. In this algorithm, features are repeatedly divided into two or more groups at each step, forming a decision tree. It is primarily employed for classification tasks, particularly with datasets containing nominal features, where the goal is to categorize data into distinct classes or categories.

```python
#Importing library foe decision classifier
from sklearn.tree import DecisionTreeClassifier
id3=DecisionTreeClassifier(criterion='entropy')
#Fit the data
k=id3.fit(x_train,y_train)
#predict the data
y_pred=id3.predict(x_test)
print(y_pred)
```

# Confusion Matrix

The confusion matrix serves as a critical tool for assessing the performance of classification models when applied to a specific test dataset. This evaluation can only be conducted when the actual values for the test data are available. The matrix itself is straightforward to interpret, although some of the associated terminology can be perplexing. Essentially, it presents the model's performance errors in a matrix format, earning it the alternative name of an error matrix.

```python
#Confusion matrix 
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_pred,y_test)
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted-y')
plt.ylabel('actual-y')
plt.show()
```

The model's accuracy score is a reflection of its capacity to accurately predict both positive and negative outcomes among all predictions. Mathematically, it signifies the ratio of the sum of true positives and true negatives to the total number of predictions made:

Accuracy Score = (TP + TN) / (TP + FN + TN + FP)

The accuracy score is calculated by comparing the predicted output value to the actual test value. On the other hand, the model's score is determined by examining the relationship between the output and input values during testing.

```python
#Accuracy score and model score
print(classification_report(y_pred,y_test))
print('accuracy-score',accuracy_score(y_pred,y_test))
print('Model score',id3.score(x_test,y_test))
```

Prediction of species

Here we can classify the three types of species of Iris flowers.

```python
X_new = np.array([[3, 2, 1, 0.2, 1], [4.9, 2.2, 3.8, 1.1, 2], [5.3, 2.5, 4.6, 1.9, 3]])

#Prediction of the species
prediction = id3.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```
